```python
# Load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Generate sample data
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
# Plot basic scatterplot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y)
```

```python
# Modify graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x, y)
```

```python
# Plotting a line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
```

```python
# Plotting a histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
# Create font dictionary
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

# Add fontdic arguments
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
# Move labels using the loc argument
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
# Remake scatterplot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,
           y,
           marker='*',
           color = "indigo")
fig.show()
```

```python
# Adjust style and width of the line connecting the points
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*', color = 'indigo', linestyle = '--', linewidth = 2)
fig.show()
```

```python
# Use hex codes to modify color
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2)
fig.show()
```

```python
# Customize plot further
ax.plot(x,y,marker='*',
        markersize = 12,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#fa9359' )
ax.grid(axis = 'y')
ax.grid(axis = 'y', color = "blue", linewidth = 2, linestyle = '-.')
```

```python
# Activity
# libraries & dataset
import seaborn as sns
df = sns.load_dataset('iris')

# set seaborn style
sns.set_style("white")

# Basic 2D density plot
sns.kdeplot(x=df.sepal_width, y=df.sepal_length)
plt.show()

# Custom the color, add shade and bandwidth
sns.kdeplot(x=df.sepal_width, y=df.sepal_length, cmap="Reds", shade=True, bw_adjust=.5)
plt.show()

# Add thresh parameter
sns.kdeplot(x=df.sepal_width, y=df.sepal_length, cmap="Blues", shade=True, thresh=0)
plt.show()
```

Recall the aesthetic, substantice, and perceptual qualities of data visualizations. Does your visualization of choice succeed in each area?

Aesthetic: This visualization is pleasing to look at as the shades are very soothing to the eye.

Substantive: The visualization accurately and honestly present data as we are not hiding or adding thresholds to our plot.

Perceptual: We can understand the message of the visulization, which is how strong the correlation of the axis lies based on where the darkest shade is located.
