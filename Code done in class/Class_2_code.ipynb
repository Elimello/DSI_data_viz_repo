```python
# Load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Generate sample data seed
np.random.seed(613)

# Generate sample data axis
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot scatter visual
ax.scatter(x, y)
```

```python
# Modify graph
fig, ax = plt.subplots(figsize=(5, 3))

# Plot bar chart
ax.bar(x, y)
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.plot(x, y)
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a histogram
ax.hist(y)
```

```python
# Create two font dictionaries
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.plot(x,y)

# Add fontdic arguments
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)

# Add layout type
fig.tight_layout()
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.plot(x,y)

# Add fontdic arguments and move labels using the loc argument
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)

# Add layout type
fig.tight_layout()
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Remake scatterplot with specific color and marker
ax.scatter(x,
           y,
           marker='*',
           color = "indigo")
fig.show()
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Adjust style and width of the line connecting the points
ax.plot(x, y, marker='*', color = 'indigo', linestyle = '--', linewidth = 2)
fig.show()
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Use hex codes to modify color
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2)
fig.show()
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Customize plot further, add marker redge, face color, and width
ax.plot(x,y,marker='*',
        markersize = 12,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#fa9359' )
ax.grid(axis = 'y')
ax.grid(axis = 'y', color = "blue", linewidth = 2, linestyle = '-.')
fig.show()
```

```python
# Import Libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a list of word
text=("Python Python Python Matplotlib DSI Data Machine Learning Science Data Analytics Business Intelligence Tech World R Statistics GitHub SQL Innovation Development Presentations Time Series DSI DSI")

# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0, background_color="grey", colormap="Greens").generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

Recall the aesthetic, substantice, and perceptual qualities of data visualizations. Does your visualization of choice succeed in each area?

**Aesthetic:** This visualization is pleasing to look at as the color shades and background are very soothing to the eye.

**Substantive:** The visualization accurately and honestly present data as we are not hiding or adding thresholds to our word cloud.

**Perceptual:** We can understand the message of the visulization, which is the bigger the word fonts and more times that word has been found in the text provided.
