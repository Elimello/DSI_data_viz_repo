```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.plot(x, y1, y2)
fig.show()
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc='lower right',
          frameon = True,
          fontsize = 12,
          ncol = 2,
          shadow = True)
fig.show()
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc='upper left',
          bbox_to_anchor=(1,1),
          frameon = True,
          fontsize = 10,
          shadow = True)
fig.show()
```

```python
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10,95, "This value is important!",
        ha='left',
        color='red',
        size=8)
fig.show()
```

```python
# Positioning text annotations
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

ax.text(1, 5, "Data: (1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, "Axes Object: (0.5,0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, "Figure: (0.2, 0.2)", transform=fig.transFigure)
```

```python
# Adding an arrow shape next to the annotation
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!',
            xy = (10,95),
            xytext = (20, 94),
            arrowprops = dict(facecolor = 'grey',color='grey', shrink = 0.05))

fig.show()
```

```python
# Removing tick marks and labels
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())

fig.show()
```

```python
# Removing tick marks and labels
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MaxNLocator(3))

fig.show()
```

```python
# Tick mark intervals
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

fig.show()
```

```python
# Rotating axis labels
# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right', fontsize = 12, font = "serif")
#ax.tick_params(axis='x', colors='grey')

fig.show()
```

```python
# Modifying axis title fonts

font1 = {'family':'sans-serif','color':'indigo','size':15}


# Declare visual specs with specific size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line chart
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right', fontsize = 12, font = "serif")
plt.xlabel('Shiny New X Axis!', fontdict = font1)

fig.show()
```
