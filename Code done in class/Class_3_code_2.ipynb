```python
# Subplots

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# Activity

plt.style.use('dark_background')

fig, (ax1, ax2) = plt.subplots(ncols = 2,
                               nrows = 1,
                               figsize = (7,3))

ax1.scatter(x1, y1)
ax2.bar(x2, y2)
plt.xticks(rotation=45, ha='right', fontsize = 12, font = "serif")
plt.yticks(rotation=45, ha='right', fontsize = 12, font = "serif")

fig.show()
```
